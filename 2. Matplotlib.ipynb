## Визуализация

### Импортируем библиотеки

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Минимализм

```python
plt.plot(1, 2, "o")
plt.plot(2, 4, "o")
plt.plot(np.array([2,2,3]), np.array([-1,3,7]), "o")
plt.plot(np.arange(1, 5),(np.arange(1, 5)*2), "-")
```

### Координатная сетка

```python
plt.figure(figsize=(5,5))

plt.xlim(-5, 10)
plt.ylim(-5, 10)
plt.grid()
plt.plot(0,0, "o", color='black')
plt.plot(np.arange(-5, 11), np.arange(-5, 11)*0, "-", color='black')
plt.plot(np.arange(-5, 11)*0, np.arange(-5, 11), "-", color='black')

plt.show()
```

### Комментарии

```python
x= np.arange(-5, 10)
y = 0.5*x
plt.plot(x,y,'-')
```

```python
x= np.arange(-5, 10)
y = 7*x + 3
plt.plot(x,y,'-')
```

```python
plt.figure(figsize=(5,5))

plt.xlim(-5, 10)
plt.ylim(-5, 10)
plt.grid()
plt.plot(0,0, "o", color='black')
plt.plot(np.arange(-5, 11), np.arange(-5, 11)*0, "-", color='black')
plt.plot(np.arange(-5, 11)*0, np.arange(-5, 11), "-", color='black')

x= np.arange(-5, 10)
y = 0.5*x
z = 7*x + 3
plt.plot(x,y,'-')
plt.plot(x,z,'-')

plt.show()
```

```python
a1 = np.array([5,6,3,2,7,8,9,1,0,4])
a2 = np.array([5,7,3,3,7,9,9,0,0,6])

plt.figure(figsize=(5,5))

plt.xlim(-5, 10)
plt.ylim(-5, 10)
plt.grid()
plt.plot(0,0, "o", color='black')
plt.plot(np.arange(-5, 11), np.arange(-5, 11)*0, "-", color='black')
plt.plot(np.arange(-5, 11)*0, np.arange(-5, 11), "-", color='black')

plt.plot(a1,a2,'.')

plt.show()
```
